# Проект 6.  Аналитика бизнес-метрик для Яндекс.Афиши

## 1. Введение

В нашем распоряжении - данные от Яндекс.Афиши за календарный год. **Цель** - помочь маркетологам оптимизировать маркетинговые затраты.<br>

**Описание данных**<br>
Таблица `visits` (лог сервера с информацией о посещениях сайта):<br>

•	Uid — уникальный идентификатор пользователя<br>
•	Device — категория устройства пользователя<br>
•	Start Ts — дата и время начала сессии<br>
•	End Ts — дата и время окончания сессии<br>
•	Source Id — идентификатор рекламного источника, из которого пришел пользователь<br>

Таблица `orders` (информация о заказах):<br>
•	Uid — уникальный id пользователя, который сделал заказ<br>
•	Buy Ts — дата и время заказа<br>
•	Revenue — выручка Яндекс.Афиши с этого заказа<br>

Таблица `costs` (информация о затратах на маркетинг):<br>
•	source_id — идентификатор рекламного источника<br>
•	dt — дата<br>
•	costs — затраты на этот рекламный источник в этот день<br>

**План исследования**<br>
Ответим на следующие вопросы:<br>

По продукту:<br>
•	Сколько людей пользуются в день, неделю, месяц?<br>
•	Сколько сессий в день?<br>
•	Сколько длится одна сессия?<br>
•	Как часто люди возвращаются?<br>
По продажам:<br>
•	Когда люди начинают покупать?<br>
•	Сколько раз покупают за период?<br>
•	Какой средний чек?<br>
•	Сколько денег приносят? (LTV)<br>
По маркетингу:<br>
•	Сколько денег потратили? Всего / на каждый источник / по времени<br>
•	Сколько стоило привлечение одного покупателя из каждого источника?<br>
•	Насколько окупились расходы? (ROI)<br>

Посмотрим, как эти метрики отличаются по устройствам и по рекламным источникам, как они меняются во времени.

## 4. Исследовательский анализ данных

### 4.1  Аналитика продукта

Добавим колонки с днем, неделей, месяцем и годом сессии.

```python
visits['ses_year'] = visits['start_ts'].dt.year
visits['ses_month'] = visits['start_ts'].dt.month
visits['ses_week'] = visits['start_ts'].dt.week
visits['ses_date'] = visits['start_ts'].dt.date
display(visits.head(2))
#visits.info()
```

Извлекая год, месяц и неделю именно таким образом (с помощью метода .dt.month и т.д.) мы получаем  значения целочисленного формата для года, месяца и недели (и формата object - для даты), с которыми можем впоследствии проводить математические операции (например, поиск среднего значения). 

#### Количество пользователей в месяц, неделю и  день (MAU, WAU, DAU)

Рассчитаем **среднее** количество уникальных пользователей в день, неделю, месяц.

```python
dau_total_mean =  int(visits.groupby('ses_date').agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей в день, DAU_MEAN:", dau_total_mean)
wau_total_mean =  int(visits.groupby(['ses_year','ses_week']).agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей в неделю, WAU_MEAN:", wau_total_mean)
mau_total_mean =  int(visits.groupby(['ses_year','ses_month']).agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей в месяц, MAU_MEAN:", mau_total_mean)
```

Посмотрим на эти же метрики, но уже в разрезе устройств.

```python
dau_touch_mean =  int(visits.query('device == "touch"').groupby('ses_date').agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей cо смартфонов и планшетов в день, DAU_TOUCH_MEAN:", dau_touch_mean)
wau_touch_mean=  int(visits.query('device == "touch"').groupby(['ses_year','ses_week']).agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей cо смартфонов и планшетов в неделю, WAU_TOUCH_MEAN:", wau_touch_mean)
mau_touch_mean =  int(visits.query('device == "touch"').groupby(['ses_year','ses_month']).agg({'uid':'nunique'}).mean().round())
print("Количество уникальных пользователей cо смартфонов и планшетов в месяц, MAU_TOUCH_MEAN:", mau_touch_mean)
```

```python
dau_desktop_mean = dau_total_mean - dau_touch_mean
wau_desktop_mean = wau_total_mean - wau_touch_mean
mau_desktop_mean = mau_total_mean - mau_touch_mean
print("Количество уникальных пользователей c компьютеров и  ноутбуков в день, DAU_DESKTOP_MEAN:", dau_desktop_mean)
print("Количество уникальных пользователей c компьютеров и  ноутбуков в неделю, WAU_DESKTOP_MEAN:", wau_desktop_mean)
print("Количество уникальных пользователей c компьютеров и  ноутбуков в месяц, MAU_DESKTOP_MEAN:", mau_desktop_mean)
```

Построим для наглядности графики.

```python
au_total_mean = [dau_total_mean, wau_total_mean, mau_total_mean]
au_desktop_mean = [dau_desktop_mean, wau_desktop_mean, mau_desktop_mean]
au_touch_mean = [dau_touch_mean, wau_touch_mean, mau_touch_mean]

print("TOTAL_MEAN:", au_total_mean)
print("DESKTOP_MEAN:", au_desktop_mean)
print("TOUCH_MEAN:", au_touch_mean)

au_mean = [au_touch_mean , au_desktop_mean, au_total_mean ]
print(au_mean)

device = ['touch','desktop','total'  ]
headings=['dau_mean', 'wau_mean', 'mau_mean']
active_users = pd.DataFrame(au_mean, index = device, columns = headings)
display(active_users)
```

```python
active_users.plot(kind = 'bar', style='o-', grid = True, figsize = (9,7), 
                   title = 'Показатели средней пользовательской активности в день, неделю, месяц (всего и в разрезе по девайсам)');
```

```python
active_users = active_users.T
active_users
```

```python
active_users['touch_%']= (100*active_users['touch']/active_users['total']).round()
active_users['desktop_%']= (100*active_users['desktop']/active_users['total']).round()
active_users
```

На графике и из расчетов хорошо видно, что ***touch-устройства сильно проигрывают*** в пользовательской активности ***desktop-устройствам: средняя активность пользователей через компьютеры и ноутбуки в 2,5 раза выше, чем через планшеты и смартфоны ( 71% против 29%)***. Причем статистика абсолютно одинакова ***как для дневной активности, так и для недельной и месячной***. Таким образом, можно порекомендовать маркетологам, разрабатывать и ***запускать больше акций, нацеленных на аудиторию desktop-устройств**, а веб-дизайнерам и разработчикам - ***обратить больше внимания и усилий на развитие desktop-интерфейса***.

#### Количество сессий в день и средняя продолжительность сессии (ASL)


Добавим столбец с продолжительностью сессии.

```python
visits['dur_ts'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
display(visits.head(3))
type(visits['dur_ts'][0]) # проверяем тип получившегося столбца - целое число
```

Посмотрим, сколько сессий нулевой продолжительностью.

```python
visits[visits['dur_ts'] == 0]['dur_ts'].count()
```

"Нулевых" сессий порядка 10%. В принципе, это укладывается в разумный процент неудавшихся (прервавшихся) сессий, хотя, для подстраховки можно обратить внимание разработчиков на этот момент: вдруг есть проблемы со входом на сайт. <br>


Отфильтруем сессии с ненулевой длительностью. Иначе мы исказим среднюю продолжительность сессии.

```python
good_visits = visits[visits['dur_ts'] > 0]
good_visits.shape[0] # кол-во ненулевых сессий
```

Проверим данные на логику. Время начала сессии не должно быть позднее времени её окончания.

```python
good_visits.query('start_ts > end_ts')
```

Таких строк две. Уберем их так же.

#### Расчет коэффициента удержания (Retention Rate)

Для того, чтобы ответить на вопрос "Как часто пользователи возвращаются?", необходимо рассчитать коэффициент удержания (Retention Rate). Для этого нам нужно разбить пользователей на когорты и посмотреть, как изменяется количество пользователей по когортам в зависимости от срока жизни когорты. Сделаем для этого ряд предварительных действий.<br>

Поскольку мы имеем данные за полный год, имеет смысл разбить пользователей на когорты по месяцам.<br>

Создадим для этого еще раз столбцы с датой и месяцем активности пользователя (сессии), но уже в формате даты (а не в числовом формате, как требовалось нам для предыдущего шага).

```python
visits['act_date'] = visits['start_ts'].dt.date
visits['act_date'] = pd.to_datetime(visits['act_date'], format = '%Y-%m-%d')

visits['act_month'] = visits['start_ts'].astype('datetime64[M]') 
display(visits.head(3))
```

Найдем дату первой активности пользователя. Добавим ее к исходному датасету путем объединения методом 'join', заодно оставим только нужные нам столбцы.

```python
first_act_date = visits.groupby('uid')['act_date'].min()
first_act_date.name = 'first_act_date'
visits_1 = visits.join(first_act_date, on = 'uid')[['device','source_id','uid','act_date','act_month','first_act_date']]
display(visits_1.head(1))
visits_1.info() # проверим, что дата и месяц у нас теперь нужного формата datetime
```

Найдем месяц первой активности.

```python
visits_1['first_act_month'] = visits_1['first_act_date'].astype('datetime64[M]')
visits_1.head(1)
```

Рассчитаем для каждой строки датафрейма время жизни (lifetime) пользователя в рамках когорты (она у нас месячная). Получившееся время сдвига в днях формата timedelta приведем к целому количеству месяцев.

```python
visits_1['cohort_lifetime'] = visits_1['act_month'] - visits_1['first_act_month']
visits_1['cohort_lifetime'] = (visits_1['cohort_lifetime'] / np.timedelta64(1,'M')).round().astype('int')
visits_1.sample(2)
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество уникальных активных пользователей на определённый «месяц жизни». Сбросим индексы для последующей обработки.

```python
cohorts = visits_1.groupby(['first_act_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display(cohorts)
```

Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждый следующий месяц.<br>
Найдём исходное количество пользователей в когорте. Возьмём их число на нулевой месяц.

```python
initial_cohort_users = cohorts[cohorts['cohort_lifetime'] == 0][['first_act_month','uid']]
initial_cohort_users.T
```

Переименуем столбец `uid` в `cohort_users`. Объединим данные по когортам с исходным количеством пользователей в когорте.

```python
initial_cohort_users = initial_cohort_users.rename(columns = {'uid':'cohort_users'})
cohorts = pd.merge(cohorts, initial_cohort_users, on = 'first_act_month' )
display(cohorts.head(3))
```

Рассчитаем Retention Rate, разделив количество активных пользователей в каждый месяц на исходное число пользователей в когорте.

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
display(cohorts.head(3))
```

Для наглядности создадим сводную таблицу и тепловую карту.

```python
cohorts['first_act_month'] = cohorts['first_act_month'].dt.strftime('%Y-%m') # уберем день, оставим только год и месяц
retention_pivot = cohorts.pivot_table(index = 'first_act_month', columns = 'cohort_lifetime', values = 'retention')
display(retention_pivot.fillna(''))
```

```python
# set seaborn heatmap title, x-axis, y-axis label and font size
 
plt.figure(figsize=(14,8))
 
ax = sns.heatmap(retention_pivot, annot = True, fmt = '.1%', linewidths=1, linecolor='gray', vmin = 0.0, vmax = 0.1);
 
ax.set(title="Коэффициент удержания (Retention Rate) по когортам",
      xlabel="Возраст когорты, в месяцах",
      ylabel="Когорты",)
 
sns.set(font_scale = 1) # set fontsize 1

#fig, ax = plt.subplots()
#ax.set_title('Коэффициент удержания (Retention Rate) по когортам')
#fig.set_figwidth(14)
#fig.set_figheight(7)
#  Добавляем подписи к осям:
#ax.set_xlabel('Возраст когорты, в месяцах')
#ax.set_ylabel('Когорты')
```

С помощью тепловой карты нам становится очевидной следующая картина: ***с первого же месяца коэффициент удержания (RR) очень резко падает (до 4 - 8%)***: максимальный его показатель для сентябрьской когорты составляет 8,5%, минимальный - 4,2% - для мартовской. В целом, ***заметен тренд на снижение RR от когорты к когорте***. То есть в течение рассматриваемого годового периода все больше людей уходило после нулевого месяца: если в начале периода через месяц из 100 оставалось 8 человек, то к концу периода - уже меньше 5ти.<br>
    Также ***характерно планомерное снижение уровня RR с возрастом когорт***, причем это ***касается практически всех когорт*** за исключением первой, июньской - в ней с началом сезонных рекламных кампаний (сентябрь-ноябрь) наблюдался некоторый возврат клиентов, но не более 1-2х человек из 100. <br>
    
 Из всего вышесказанного можно сделать вывод  том, что ***предоставляемые сервисом услуги покупатели рассматривают скорее как "одноразовые", то есть их не устраивает либо качество или удобство сервиса, либо его его цена***. Вопрос на проработку как маркетологам, так и владельцам продукта и веб-архитекторам и дизайнерам.

#### Сколько денег приносят покупатели (LTV - Life Time Value) 

В таблице `orders_1` мы уже получили месяц заказа и месяц первого заказа.

```python
orders_1.head(15)
```

Посчитаем количество новых покупателей (n_buyers) для каждой когорты в каждый месяц жизни когорты. 

```python
cohort_sizes = orders_1.groupby(['first_order_month','order_month']).agg({'uid': 'nunique'}).reset_index()
cohort_sizes = cohort_sizes.rename(columns = {'uid': 'n_buyers'})
display(cohort_sizes.head(20))
```

Посчитаем какую выручку принесла каждая когорта в каждом месяце.

```python
cohorts_revenue = orders_1.groupby(['first_order_month','order_month']).agg({'revenue':'sum'}).reset_index()
cohorts_revenue
```

Добавим к этой таблице столбец с количеством новых покупателей в каждой когорте.

```python
report = cohorts_revenue.merge(cohort_sizes, on = ['first_order_month','order_month'])
report
```

LTV считают не из выручки, а из валовой прибыли, то есть умножают выручку на маржинальность. Однако, в нашем проекте информации по размеру маржинальности нет. Поэтому мы делаем предположение о 100%-ной маржинальности. Посчитаем валовую прибыль (gross profit) для каждого месяца каждой когорты.

```python
margin_rate = 1.0
report['gp'] = report['revenue']* margin_rate
```

Поскольку LTV — относительный показатель, его удобнее смотреть по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты.

```python
report['age'] = (report['order_month'] - report['first_order_month'])/np.timedelta64(1,'M')
report['age'] = report['age'].round().astype('int')
report.head()
```

Найдём LTV. Разделим валовую прибыль когорт за каждый месяц на число пользователей в каждой когорте в каждый месяц ее жизни. Представим это в виде сводной таблицы.

```python
test_df = pd.DataFrame(np.array([[1, 1, 3], [1, 2, 3], [1, 2, 3],[1, 2, 1], [1, 3, 1], [1, 2, 1]]),
                   columns=['a', 'b', 'c'])
display(test_df)
test_df.groupby(['a','b']).agg({'c':'nunique'})
```

```python
report['ltv'] = report['gp']/report['n_buyers']
report
```

```python
output_ltv = report.pivot_table( index = 'first_order_month', columns = 'age',
                                values = 'ltv', aggfunc = 'sum', margins = True).round(2)  
# в качестве aggfunc возможна почти любая аггрегирующая функция (кроме first, last, min, max), например, mean,
# т.к. в каждой когорте и возрасте по одному значению
output_ltv = output_ltv.rename(columns ={'All':'ltv_total'})
output_ltv.fillna('')                      
```

Построим тепловую карту для данной сводной таблицы. 

```python
# set seaborn heatmap title, x-axis, y-axis label and font size
plt.figure(figsize=(14,8))
ax = sns.heatmap(output_ltv, annot = True, fmt = '.2f', linewidths=1, linecolor='gray', vmin = 4.0, vmax = 48.2);
 
ax.set(title="Пожизненная ценность клиента (LTV) с учетом разбивки по когортам",
      xlabel="Возраст когорты, в месяцах",
      ylabel="Когорты")
 
sns.set(font_scale = 1) # set fontsize 1
```

В итоге мы нашли LTV покупателя для каждого возраста каждой когорты. А также в столбце "ltv_total" мы видим итоговую накопленную LTV покупателя каждой когорты. Можно заметить, что ***наибольшую ценность имеют клиенты сентябрьской когорты : всего они принесли 281,64 единицы валовой прибыли на покупателя***, несмотря на то, что это не самая "старшая" когорта. Такая сумма обусловлена большим всплеском LTV на 3й месяц жизни когорты (age = 3, LTV = 138 ед), это декабрь - предновогоднее время, высокий сезон. ***На втором месте находится июньская когорта (всего 259 ед. валовой прибыли на пользователя)***, что уже вполне логично, поскольку это самая "старшая" когорта, и она успела сгенерировать денег больше, чем более "младшие". В этой когорте заметны всплески на 3й, 6й, 9й и 10й месяцы жизни (это сентябрь - начало сезона, декабрь - высокий сезон, март - женский праздник, и апрель).<br> 
Также ***выделяется декабрьская когорта - у нее заметны всплески LTV на 2й и 3й месяцы жизни*** ( это февраль и март - месяцы мужского и женского праздников).<br>
Кроме данных трех когорт по остальным когортам тенденция такова, что LTV по когортам падает, что естественно, поскольку чем позднее создана когорта, тем меньше она успела принести дохода.

#### Стоимость привлечения одного покупателя

Обратимся к ранее полученному нами отчету `report`.

```python
report.head(2)
```

Добавим в него столбец с посчитанными месячными затратами. И найдем CAC (стоимость привлечения покупателя) для каждой когорты.

```python
revenue_costs = pd.merge(report, monthly_costs, left_on = 'first_order_month', right_on = 'month')
revenue_costs['cac'] = revenue_costs['costs']/revenue_costs['n_buyers']
revenue_costs.head()
```

Теперь найдем стоимость привлечения одного покупателя для каждого источника. Уже имеем разбивку затрат по месяцам и по источникам `source_monthly_costs`.

```python
source_monthly_costs = source_monthly_costs.reset_index()
source_monthly_costs.head(3)
```

Вспомним про объединенную таблицу визитов и заказов `orders_source`, где каждому покупателю мы сопоставили именно его первый источник (источник первой сессии).

```python
orders_source['first_act_month'] = orders_source['first_act_time'].astype('datetime64[M]')
orders_source.head(1)
```

Сгруппируем таблицу по когортам (месяцу первой активности) и по источнику, и посчитаем количество новых покупателей в каждой группе. Переименуем получившийся столбец и сбросим индексы.

```python
new_buyers_cohorts_by_source = orders_source.groupby(['first_act_month','source_id'])  \
                                                            .agg({'uid':'nunique'}).reset_index()
new_buyers_cohorts_by_source = new_buyers_cohorts_by_source.rename(columns = {'uid':'new_buyers'})
new_buyers_cohorts_by_source.head(3)
```

#### Насколько окупились расходы (ROI)

```python
orders_source.head(3)
```

Посчитаем выручку по когортам в разрезе источников.

```python
cohorts_source_revenue = orders_source.groupby(['first_act_month','source_id']).agg({'revenue':'sum'}).reset_index()
cohorts_source_revenue.head()
```

Объединим с таблицей с данными по затратам. Посчитаем ROI (возврат на инвестиции). Представим показатель в процентах и округлим до 1 знака после запятой.

```python
revenue_costs_buyers = pd.merge(cohorts_source_revenue, costs_buyers,
                                on = ['first_act_month', 'source_id'], how = 'inner')
revenue_costs_buyers['roi'] = (
    ((revenue_costs_buyers['revenue'] - revenue_costs_buyers['costs'])/ revenue_costs_buyers['costs'])*100)
revenue_costs_buyers = revenue_costs_buyers.round(1)
revenue_costs_buyers
```

```python
roi_pivot = revenue_costs_buyers.pivot_table(index = 'first_act_month',columns = 'source_id', values = 'roi')
roi_pivot 
```

```python
roi_pivot.plot(style = 'o-', figsize = (14,9) , grid = True,
                   title = 'ROI по когортам в разрезе рекламных источников');
```

   Как видно из графика, в начале рассматриваемого года ***(июнь 2017г.) окупались затраты почти во все источники***, за исключением 3-го и 10-го. Причем ***окупаемость составляла от 1,5 до 2,5 раз, а источник №1 показал даже окупаемость почти в 700%***. Затем практически по всем источником пошло снижение ROI (кроме сентябрьского всплеска по источнику №5) и ***к декабрю все источники (за исключением 2-го, он показал скачок до окупаемости инвестиций почти в 200%) перестали окупаться***. И до конца рассматриваемого периода (до мая 2018 г.) все источники так и остались в "минусовой" зоне окупаемости маркетинговых затрат. То есть ***резкое увеличение маркетинговых вливаний в высокий сезон (с декабря) себя не оправдало***.

Теперь посмотрим, сколько месяцев нужно каждой когорте, чтобы окупить маркетинговые затраты. <br>
Обратимся к таблице `revenue_costs`, где у нас уже посчитаны LTV  и CAC.

```python
revenue_costs.head(3)
```

Рассчитаем ROMI (возврат на маркетинговые инвестиции), разделив LTV на CAC. Построим сводную таблицу. Применим метод `cumsum` с параметром `axis = 1` , чтобы построчно посчитать накопительный (кумулятивный) ROMI для каждой когорты. И построим тепловую карту.

```python
revenue_costs['romi'] = revenue_costs['ltv']/revenue_costs['cac']

# приведем столбец с месяцем создания когорты к удобному для визуального представления виду:
revenue_costs['first_order_month'] = revenue_costs['first_order_month'].dt.strftime('%Y-%m')

romi_pivot = revenue_costs.pivot_table(index = 'first_order_month', columns = 'age', values = 'romi')
romi_cumulative = romi_pivot.cumsum(axis=1).round(2)
display(romi_cumulative.fillna(''))
```
